```python
import pandas as pd
import numpy as np

file_to_load = "Resources/purchase_data.csv"

# Read File and store into Pandas data frame
game = pd.read_csv(file_to_load)
game.head()
```

```python
#Player Count
totalplayers = len(game["SN"].unique())
firstchart = pd.DataFrame({"Total players": [totalplayers]})
firstchart
```

```python
#Summary table
unique_items = len(game["Item ID"].unique())
avg_price = np.round(game["Price"].mean(), decimals=2)
totalpurchases = len(game["Purchase ID"])
totalrevenue = game["Price"].sum()

secondchart = pd.DataFrame({
    "Number of Unique Item IDs": [unique_items],
    "Average Price": [avg_price],
    "Number of Total Purchases": [totalpurchases],
    "Total Revenue": [totalrevenue]
})
secondchart["Average Price"] = secondchart["Average Price"].map("${:,.2f}".format)
secondchart["Total Revenue"] = secondchart["Total Revenue"].map("${:,.2f}".format)
secondchart
```

```python
#Gender Count
UniquePlayers = game.drop_duplicates(['SN','Gender'])
gender = UniquePlayers["Gender"].value_counts()
gender_percentage = np.round(((gender)/(len(UniquePlayers)) * 100), decimals=2)
chart3 = pd.concat([gender, gender_percentage], axis=1)
chart3.columns = ["Total Count", "Percentage of Players"]
chart3
```

```python
# Purchasing Analysis (Gender)
gga = game[["Purchase ID", "SN", "Gender", "Price"]]

ggdeb = gga.drop_duplicates(['SN','Gender'])
ggdebb = ggdeb.groupby(["Gender"])
gghy = ggdebb.size()

ggb = gga.groupby(["Gender"])

#tpv = total purchase value $
ggbtpv = ggb["Price"].sum()
#pc = Purchase Count
ggbpc = ggb["Purchase ID"].count()
#avg price
ggbavg = np.round((ggbtpv / ggbpc), decimals=2)
#avg total per person
ggbatpp = np.round((ggbtpv / gghy), decimals=2)

chart4 = pd.concat([ggbpc, ggbavg, ggbtpv, ggbatpp], axis=1)

chart4.columns = ["# of Purchases", "Price of Average Purchase", "Total Purchase Value", "Average Total Purchase per Person"]
chart4["Price of Average Purchase"] = chart4["Price of Average Purchase"].map("${:,.2f}".format)
chart4["Average Total Purchase per Person"] = chart4["Average Total Purchase per Person"].map("${:,.2f}".format)
chart4["Total Purchase Value"] = chart4["Total Purchase Value"].map("${:,.2f}".format)

chart4
```

```python
# Age Demographics
bins2 = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names2 = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

AG = game[["SN", "Age"]]
AG = AG.drop_duplicates(['SN'])
AG.loc[:,"Age Bracket"] = pd.cut(AG["Age"], bins2, labels=group_names2)
AG2 = AG.groupby("Age Bracket")
ag2_pop = pd.DataFrame(AG2["Age"].count())
ag2_avg = np.round(ag2_pop/len(AG["Age"]) * 100, decimals=2)
chart13 = pd.concat([ag2_pop, ag2_avg], axis=1)
chart13.columns = ["Total Count", "Percentage of Players"]
chart13
```

```python
# Purchasing Analysis by Age

bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Gga1 = game[["Purchase ID", "SN", "Age", "Price"]]
Gga1.loc[:,"Age Bracket"] = pd.cut(Gga1["Age"], bins, labels=group_names)

UP2 = Gga1.drop_duplicates(['SN'])
UP2G = UP2.groupby("Age Bracket")
UP2Gs = UP2G.size()

Ggb1 = Gga1.groupby(["Age Bracket"])

#tpv = total purchase value $
adtpv = Ggb1["Price"].sum()
#pc = Purchase Count
adpc = Ggb1["Purchase ID"].count()
#avg price
adavg = np.round((adtpv / adpc), decimals=2)
#avg total per person
adatpp = np.round((adtpv / UP2Gs), decimals=2)

chart5 = pd.concat([adpc, adavg, adtpv, adatpp], axis=1)

chart5.columns = ["# of Purchases", "Price of Average Purchase", "Total Purchase Value", "Average Total Purchase per Person"]
chart5["Price of Average Purchase"] = chart5["Price of Average Purchase"].map("${:,.2f}".format)
chart5["Average Total Purchase per Person"] = chart5["Average Total Purchase per Person"].map("${:,.2f}".format)
chart5["Total Purchase Value"] = chart5["Total Purchase Value"].map("${:,.2f}".format)

chart5
```

```python
# Top Spenders(TS)

TS = game[["SN", "Price"]]
TS2 = TS.groupby("SN")

# pc is purchase count
# app is average purchase price
# tpv is sum of total purchases
TS2pc = TS2.size()
TS2app = np.round((TS2["Price"].mean()), decimals=2)
TS2tpv = TS2["Price"].sum()

TS3 = pd.concat([TS2pc, TS2app, TS2tpv], axis=1)

TS3.columns = ["Purchase Count", "Price of Average Purchase", "Total Purchase Value"]
TS4 = TS3.sort_values(["Total Purchase Value"], ascending=False)
TS4["Price of Average Purchase"] = TS4["Price of Average Purchase"].map("${:,.2f}".format)
TS4["Total Purchase Value"] = TS4["Total Purchase Value"].map("${:,.2f}".format)

TS4.head(5)
```

```python
# Most Popular Items (MPI)

MPI = game[["Item ID", "Item Name", "Price"]]
MPI2 = MPI.groupby(["Item ID", "Item Name"])

# pc is purchase count
# ip is item price
# tpv is sum of total purchases
MPI2pc = MPI2.size()
MPI2app = np.round((MPI2["Price"].mean()), decimals=2)
MPI2tpv = MPI2["Price"].sum()

MPI3 = pd.concat([MPI2pc, MPI2app, MPI2tpv], axis=1)

MPI3.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
MPI4 = MPI3.sort_values(["Purchase Count"], ascending=False)
MPI4["Item Price"] = MPI4["Item Price"].map("${:,.2f}".format)
MPI4["Total Purchase Value"] = MPI4["Total Purchase Value"].map("${:,.2f}".format)

MPI4.head(5)
```

```python
# Most Profitable Items

MPI5 = MPI3.sort_values(["Total Purchase Value"], ascending=False)
MPI5["Item Price"] = MPI5["Item Price"].map("${:,.2f}".format)
MPI5["Total Purchase Value"] = MPI5["Total Purchase Value"].map("${:,.2f}".format)

MPI5.head()
```
